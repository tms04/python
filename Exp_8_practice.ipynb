```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,5))
plt.plot([1,2,3,4],[1,4,9,16],"go",np.arange(1,5),np.arange(1,5)**3,"r^")
plt.title("First Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.title("1st graph")
plt.plot([1,2,3,4],[1,4,9,16],"go")

plt.subplot(2,2,4)
plt.title("2nd graph")
plt.plot(np.arange(1,5),np.arange(1,5)**3,"r^")

plt.suptitle("Graphs")
plt.show()
```

```python
x=np.arange(1,5)
y=x**3
fig, ax=plt.subplots(2,2,figsize=(6,6))
ax[0,1].plot([1,2,3,4],[1,4,9,16],"go")
ax[0,1].set_title("1st graph")
ax[1,0].plot(x,y,"r^")
ax[1,1].set_title("2nd graph")
plt.show()
```
